```python
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import classification_report, accuracy_score
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv('E:/aleksa_praksa/nlp_internship/data-internship/data/tokenized_dataset.csv')
```

```python
df.head()
```

```python
import ast

df['tokens'] = df['tokens'].apply(ast.literal_eval) # str to list
```

```python

df['text_joined'] = df['tokens'].apply(lambda x: " ".join(x))
```

```python

df['text_joined'] = df['tokens'].apply(lambda x: " ".join(x))
```

```python
df.head()
```

```python
language_mapping = {
    'en': 0,
    'de': 1,
    'fr': 2,
    'es': 3
}
```

```python
df['encoded_lang'] = df['language'].map(language_mapping)
```

```python
# English
X_train, X_test, y_train, y_test = train_test_split(df['text_joined'], df['encoded_lang'], test_size = 0.2, stratify = df['language'], random_state = 42)
```

```python
vectorizer = TfidfVectorizer(tokenizer=str.split,preprocessor=None,lowercase=False)
```

```python
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec  = vectorizer.transform(X_test)
```

```python
lr = LogisticRegression()
```

```python
param_grid_logreg = {
    "C": [0.1, 1, 3],
    "penalty": ["l2"],
    "solver": ["liblinear"], 
    'max_iter': [200,300]    
}
```

```python
grid = GridSearchCV(estimator=lr, param_grid=param_grid_logreg, cv=3, scoring="accuracy", n_jobs=-1, verbose=1)
```

```python
print(y_train)
```

```python
grid.fit(X_train_vec, y_train)
```

```python
print(f"\nBest params for LR: {grid.best_params_}")
print(f"Best CV accuracy: {grid.best_score_:.4f}\n")
```

```python
best_params = grid.best_params_
best_model = LogisticRegression(**best_params)
best_model.fit(X_train_vec, y_train)
```

```python
preds = best_model.predict(X_test_vec)
```

```python
print(classification_report(y_test, preds))
```

```python
acc = accuracy_score(y_true = y_test, y_pred = preds)
```

```python
class_report = classification_report(y_test, preds, output_dict=True)
class_report
```

```python
import mlflow
import json
import pickle
```

```python
mlflow.set_tracking_uri("http://127.0.0.1:5000/")
experiment_name = "Language_Prediction2"
mlflow.set_experiment(experiment_name)
```

```python
with mlflow.start_run(run_name='LogisticRegression'):
    for p, v in grid.best_params_.items():
        mlflow.log_param(p, v)

    mlflow.log_metric('test_accuracy', acc)

    # Save classification report as JSON
    report_json = json.dumps(class_report, indent=2)
    mlflow.log_text(report_json, f"LogisticRegression_classification_report.json")

    mlflow.sklearn.log_model(sk_model = best_model, name = 'LogisticRegression_lang_pred_model')

```
